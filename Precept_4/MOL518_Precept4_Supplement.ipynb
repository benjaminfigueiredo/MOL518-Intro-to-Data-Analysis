# USGS data query

To get this data, I have queried [the US Geological Survey's earthquake database](https://earthquake.usgs.gov/earthquakes/search/) for all earthquakes of magnitude over 3 in the Richter scale recorded between 2021-01-01 00:00:00 and 2026-02-18 23:59:59, with epicenters located in the [16.254, 50.929] latitude range and [-130.254, -100.02] longitude range, which encompasses the surrounding of the San Andreas fault. The result of this query has been saved in data/query.csv

# Creating a numpy array from the query

```python
datapath = 'data/query.csv'
import pandas as pd
# read the csv file into a pandas dataframe
df = pd.read_csv(datapath)
# we will now change the time column to a datetime object and then a numeric value for processing
df['time'] = pd.to_datetime(df['time']) 
df = df.sort_values('time') # we sort the dataframe by time just in case it is not sorted
df['time_numeric'] = df['time'].astype('int64') // 10**9  # convert to seconds since Unix epoch
# we will save the numeric time column, alongside the magnitude to a numpy array
import numpy as np
data = df[['time_numeric', 'mag']].values
# save this data to a new csv file
np.savetxt('data/earthquake.csv', data, delimiter=',', header='time,magnitude', comments='')
```
